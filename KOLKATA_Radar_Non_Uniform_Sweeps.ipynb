```python
import os
import shutil

def load_files_from_dir(raw_dir):
    file_list = os.listdir(raw_dir)
    return file_list

def process_file(file_path, organized_dir):
    ds = xr.open_dataset(file_path)
    sweep_number = ds.dims['sweep']
    
    # Create the directory
    directory_name = f"sweep_{sweep_number}"
    os.makedirs(os.path.join(organized_dir, directory_name), exist_ok=True)
    
    # Move the file to the directory
    destination_path = os.path.join(organized_dir, directory_name, os.path.basename(file_path))
    shutil.move(file_path, destination_path)
    
    print(f"File '{file_path}' moved to '{destination_path}'")
```

```python
# Example usage
raw_dir = "./CAL220430IMD-C/"
organized_dir = "./CAL_organized/"
files = load_files_from_dir(raw_dir)

# Create 'organized' directory if it doesn't exist
os.makedirs(organized_dir, exist_ok=True)

for file in files:
    file_path = os.path.join(raw_dir, file)
    process_file(file_path, organized_dir)
```

```python
def load_folder_names(organized_dir):
    folder_names = next(os.walk(organized_dir))[1]
    return folder_names

# Example usage
folders = load_folder_names(organized_dir)
print(folders)
```

```python
input_dir = organized_dir
output_dir = "./CAL_out_directory/"

folders = load_folder_names(input_dir)

for folder_name in folders:
    folder_path = os.path.join(input_dir, folder_name)
    sweep_number = int(folder_name.split("_")[1])
    pcf.cfrad(input_dir=folder_path, output_dir=output_dir, scan_type="C", nf=sweep_number)
```

```python
import numpy as np

def mask_data(radar, field, gatefilter):
    print(f'Masking: {field} \n')
    dsp = pyart.correct.despeckle_field(
        radar, field, gatefilter=gatefilter, size=10)
    new_field = radar.fields[field].copy()
    new_field['data'] = np.ma.masked_where(
        dsp.gate_included == False, radar.fields[field]['data'])
    radar.add_field(field, new_field, replace_existing=True)
    return radar

def filter_data(radar, refl_field, vel_field,
                ncp_field=None, ncp_thresh=None,
                rho_field=None, rho_thresh=None,):
    '''Remove noise based on velocity texture,snr, and rhohv, and mask all the fields'''
    texture = pyart.retrieve.calculate_velocity_texture(radar,
                                                        vel_field=vel_field,
                                                        wind_size=3,
                                                        check_nyq_uniform=False)
    radar.add_field('VT', texture, replace_existing=True)
    gatefilter = pyart.filters.GateFilter(radar)
    gatefilter.include_inside(refl_field, 10, 80)
    gatefilter.exclude_above("VT", 5)
    
#     if ncp_thresh is not None:
#         gatefilter.exclude_below(ncp_field, ncp_thresh)
#     else:
#         pass
    
#     if rho_thresh is not None:
#         gatefilter.exclude_below(rho_field, rho_thresh)
#     else:
#         pass
    
    radar = mask_data(radar, refl_field, gatefilter)
    radar = mask_data(radar, vel_field, gatefilter)
    radar.scan_type = 'ppi'
    return radar
```

```python
def get_shape(max_horz_lim_km=220, max_vert_lim_km=10, horz_res=1000, vert_res=500):
    '''
    Get target grid shape
    
    Parameters
    ----------
    max_horz_lim_km: Maximum Horizontal Length from Radar in km
    max_vert_lim_km: Maximum Vertical Length from ground in km
    horz_res: Horizontal resolution of a grid in meters
    vert_res: Vertical resolution of a grid in meters
    '''
    hshp = max_horz_lim_km * 1e3 / horz_res
    vshp = max_vert_lim_km * 1e3 / vert_res
    shape = (int(vshp+1), int(hshp+1), int(hshp+1))
    print(f"Shape: {shape}")
    return shape
```

```python
grid_shape = get_shape(max_horz_lim_km=220, max_vert_lim_km=10, horz_res=1000, vert_res=500)
```

```python
def write_grid(radar, outdir="./"):
    max_rng = 220 * 1e3 # in meters
    grid = pyart.map.grid_from_radars(radar, grid_shape,
                                      ((0., 10e3), (-max_rng, max_rng),
                                       (-max_rng, max_rng)),
                                      weighting_function='Barnes2',
                                      fields=["REF"])
    os.makedirs(outdir, exist_ok=True)
    print(f"Saving in: {outdir} as {os.path.basename(file)}\n")
    # Write grid
    pyart.io.write_grid(filename=os.path.join(outdir, f'grid_{os.path.basename(file)}'), grid=grid)
```
